```python
import numpy  as np
import  pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("hello")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
#subplot
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,4)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)

plt.plot(x,y)
```

```python
#control the canvas means overlap one into another 
```

```python
fig = plt.figure(figsize = (20,8),dpi =100)

axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.3,0.5,0.5])
axis2.plot(x,y)
```
